```python
from scipy.optimize import curve_fit
import numpy as np

def fit_mf(ydata):
    def func(x, a, b, c, d, e):
        return a * np.exp(-(x-b)**2/(2*c**2)) + d*x+e

    a = np.max(y)
    b = np.mean(y)
    c = np.std(y)
    d = 0
    e = 0
    try:
        popt, pcov = curve_fit(func, xdata, ydata, (a, b, c, d, e))
    except:
        a = np.max(y)
        b = np.mean(y)
        c = 1e-5
        d = 0
        e = 0
        popt, pcov = curve_fit(func, xdata, ydata, (a, b, c, d, e))
        
    return popt
```
